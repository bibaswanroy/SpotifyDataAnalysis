```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
#checking for nulls 
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
#10 Least Popular songs
least_popular = df_tracks.sort_values('popularity', ascending=True).head(10)
least_popular
```

```python
df_tracks.describe().transpose()
```

```python
#top 10 songs with popularity > 90
more_than_90 = df_tracks[df_tracks['popularity'] > 90].sort_values('popularity', ascending=False).head(10)
more_than_90
```

```python
df_tracks.set_index('release_date', inplace=True)
```

```python
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.index
```

```python
df_tracks[['artists']].iloc[18]
```

```python
#ms to s
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x:round(x/1000))
df_tracks
```

```python
corr_df = df_tracks.drop(['key','explicit','mode'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True,fmt='.1g',vmin=-1,vmax=1,cmap='inferno',linewidths=1, center=0,linecolor='Black')
heatmap.set_title('Correlation HeatMap between Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
sample_df
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x='energy',y='loudness').set(title="Loudness vs Energy Correlation")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x='acousticness',y='popularity',color='b').set(title="Popularity vs Acousticness Correlation")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks['dates'] = pd.to_datetime(df_tracks['dates'])
year = df_tracks.dates.dt.year
```

```python
sns.displot(year,discrete=True,height=5,aspect=2,kind='hist').set(title='Songs per Year')
```

```python
total_duration = df_tracks.duration
plt.figure(figsize=(18,7))
sns.barplot(x=year,y=total_duration,errwidth=False).set(title='Duration vs Year')
plt.xticks(rotation=90)
plt.show()
```

```python
total_duration = df_tracks.duration
plt.figure(figsize=(18,7))
sns.lineplot(x=year,y=total_duration).set(title='Year vs Duration')
plt.xticks(rotation=60)
plt.show()
```
